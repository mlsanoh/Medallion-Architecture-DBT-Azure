```python

# Configuration de la clé d'accès pour le compte de stockage Azure Data Lake Storage Gen2

spark.conf.set(
    "fs.azure.account.key.stcontosoanalytics.dfs.core.windows.net",
    dbutils.secrets.get(scope="databricksScope", key="storage-key")
)

# Création des widgets pour paramétrer le notebook
dbutils.widgets.text("fileName", "")
dbutils.widgets.text("table_schema", "")
dbutils.widgets.text("table_name", "")

# Récupération des paramètres
fileName = dbutils.widgets.get("fileName")
tableSchema = dbutils.widgets.get("table_schema")
tableName = dbutils.widgets.get("table_name")

# Création du schéma/base de données si elle n'existe pas
spark.sql(f"""
CREATE DATABASE IF NOT EXISTS {tableSchema}
""")

# Chemin du fichier Parquet dans la couche Bronze du Data Lake
path_parquet = (
    f"abfss://bronze@stcontosoanalytics.dfs.core.windows.net/"
    f"{fileName}/{tableSchema}.{tableName}.parquet"
)

# Création de la table externe dans le catalogue Databricks
create_table_query = f"""
CREATE TABLE IF NOT EXISTS {tableSchema}.{tableName}
USING PARQUET
LOCATION '{path_parquet}'
"""

print(create_table_query)
spark.sql(create_table_query)
```
